Implementation of K-Means

Step 1: Import Necessary Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

Generate Synthetic Data

```python
# Generate synthetic data with 3 clusters
data, _ = make_blobs(n_samples=300, centers=3, cluster_std=1.0,
random_state=42)
# Plot the data
plt.scatter(data[:, 0], data[:, 1], s=50, color='gray')
plt.title("Unclustered Data")
plt.show()
```

Step 3: Manual Implementation of K-Means

```python
# Number of clusters
k = 3
# Number of iterations
max_iterations = 100
# Randomly initialize centroids
np.random.seed(42)
centroids = data[np.random.choice(data.shape[0], k, replace=False)]
print("Initial Centroids:\n", centroids)
```

Assign Points to Clusters

```python
def assign_clusters(data, centroids):
 distances = np.sqrt(((data - centroids[:, np.newaxis])**2).sum(axis=2))
 return np.argmin(distances, axis=0)
```

Update Centroids

```python
def update_centroids(data, labels, k):
 return np.array([data[labels == i].mean(axis=0) for i in range(k)])
```

Iterative Optimization


```python
for iteration in range(max_iterations):
    # Assign points to clusters
    labels = assign_clusters(data, centroids)
    
    # Update centroids
    new_centroids = update_centroids(data, labels, k)
    
    # Check for convergence
    if np.all(centroids == new_centroids):
        print(f"Converged at iteration {iteration}")
        break
    
    centroids = new_centroids
```

```python
for i in range(k):
    plt.scatter(data[labels == i][:, 0], data[labels == i][:, 1], s=50, label=f"Cluster {i+1}")
plt.scatter(centroids[:, 0], centroids[:, 1], s=200, c='red', marker='x', label='Centroids')
plt.title("K-Means Clustering")
plt.legend()
plt.show()
```

```python
from sklearn.cluster import KMeans
# Initialize K-Means
kmeans = KMeans(n_clusters=k, init='k-means++', max_iter=300, n_init=10,
random_state=42)
labels = kmeans.fit_predict(data)
# Visualize results
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis', s=50)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200,
c='red', marker='x', label='Centroids')
plt.title("K-Means with Scikit-Learn")
plt.legend()
plt.show()
```

Importing libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
from sklearn.decomposition import PCA
import seaborn as sns
```

Load the 2024 Data

```python
# Load the 2024 CSV file
data_2024 = pd.read_csv('HappinessReport2024.csv')

# Print column names and number of data points
print("2024 Happiness Report CSV:")
print("Column Names:", data_2024.columns)
print("Number of Data Points:", len(data_2024))
```

 View data types of each column

```python

print("Data Types of Columns:")
print(data_2024.dtypes)
```

Check for missing values in the dataset

```python

missing_values_2024 = data_2024.isnull().sum()
print("Missing Values in 2024 Dataset:")
print(missing_values_2024)
```

Drop rows with any missing values

```python
data_2024 = data_2024.dropna()
```

Visualizing the pair plots

```python
# Corrected features list based on 2024 dataset column names
features_2024 = ['Explained by: Log GDP per capita', 'Explained by: Social support', 
                 'Explained by: Healthy life expectancy', 'Explained by: Freedom to make life choices', 
                 'Explained by: Generosity', 'Explained by: Perceptions of corruption', 'Dystopia + residual']

# Visualize relationships between different columns using pairplot
sns.pairplot(data_2024[features_2024])
plt.show()
```

Finding the correlation matrix 

```python
correlation_matrix_2024 = data_2024[features_2024].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix_2024, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Matrix (2024 Happiness Report)")
plt.show()
```

Standardize Features

```python
scaler = StandardScaler()
data_2024[features_2024] = scaler.fit_transform(data_2024[features_2024])
# Optionally, you can print the first few rows to verify
print(data_2024.head())
```

Elbow Method to Determine Optimal k for 2024 Data

```python

# Apply the Elbow Method to find the optimal k
inertia = []
k_values = range(1, 11)  # Test k values from 1 to 10

for k in k_values:
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
    kmeans.fit(data_2024[features_2024])
    inertia.append(kmeans.inertia_)

# Plotting the Elbow Method
plt.figure(figsize=(8, 6))
plt.plot(k_values, inertia, marker='o', linestyle='--')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.show()
```

Apply K-Means Clustering on 2024 Data

```python

k = 4  #  k is based on the Elbow Method 

kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
kmeans_labels = kmeans.fit_predict(data_2024[features_2024])

# Add the cluster labels to the dataset
data_2024['Cluster'] = kmeans_labels
```

Visualize Clusters using PCA (Principal Component Analysis)

```python
# Perform PCA to reduce dimensions to 2
pca = PCA(n_components=2)
principal_components = pca.fit_transform(data_2024[features_2024])

# Plot the clusters
plt.figure(figsize=(8, 6))
plt.scatter(principal_components[:, 0], principal_components[:, 1], c=kmeans_labels, cmap='viridis')
plt.title('K-Means Clustering with PCA')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Cluster')
plt.show()
```

Apply DBSCAN Clustering on 2024 Data

```python
# Apply DBSCAN with arbitrary eps and min_samples
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(data_2024[features_2024])

# Add the DBSCAN labels to the dataset
data_2024['DBSCAN_Cluster'] = dbscan_labels

# Plot the DBSCAN clustering results
plt.figure(figsize=(8, 6))
plt.scatter(principal_components[:, 0], principal_components[:, 1], c=dbscan_labels, cmap='viridis')
plt.title('DBSCAN Clustering with PCA')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Cluster')
plt.show()
```

Analyze Socio-Economic Indicators for Each Cluster

```python
# Group by the K-Means clusters and calculate the mean of each socio-economic indicator
cluster_analysis = data_2024.groupby('Cluster')[features_2024].mean()
print("Cluster Analysis (Socio-Economic Indicators):")
print(cluster_analysis)

```

For DBSCAN, analyze the socio-economic indicators for DBSCAN clusters

```python
dbscan_analysis = data_2024.groupby('DBSCAN_Cluster')[features_2024].mean()
print("\nDBSCAN Cluster Analysis (Socio-Economic Indicators):")
print(dbscan_analysis)
```

<B>Are there surprising countries in a cluster? Why might they be grouped this way?</B><br>
Occasionally, countries with seemingly different socio-economic statuses may appear in the same cluster. This could result from specific overlapping features, such as high generosity or similar levels of perceived corruption, despite differences in GDP or life expectancy. For instance, a middle-income country with strong social support might share a cluster with a high-income country due to comparable indicators in other dimensions.

<B>How do the clusters identified by DBSCAN differ from those by K-Means?</B><br>
DBSCAN often identifies fewer, more arbitrarily shaped clusters and can classify some points as outliers (assigned a -1 label), unlike K-Means, which always assigns every data point to a cluster. DBSCAN’s clusters may better capture irregular relationships in the data, such as countries with unique socio-economic profiles that don’t fit into any single group defined by K-Means.

<B>Which algorithm produces clusters that better align with regional or socio-economic patterns in the dataset?</B><br>
The choice of the better algorithm depends on the dataset's characteristics. K-Means tends to create clusters with clear, uniform separation, which may align with regional patterns like wealth or governance. On the other hand, DBSCAN's ability to detect outliers and irregularly shaped clusters might better reflect socio-economic patterns that don’t conform to strict boundaries, such as global socio-economic disparities or unique country profiles. A deeper comparison of the cluster analysis with real-world data is necessary to determine which approach aligns more accurately.
